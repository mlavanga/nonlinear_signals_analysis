## Logistic MAP: Phase space

The second example of strange attractor is the logistic map, which is defined by the set of equations: 
$$ x_{t+1} = x_{t}*(r-r*x_{t}) $$.

After we defined this equation in the following code block, we can iterate it in a for loop and plot the results in a Phase-Space.

```python
#python3
# import numpy as np 
# import matplotlib.pyplot as plt 
# plt.style.use('dark_background')

def logistic_attractor(x, r=4.0):
    '''Computes the next step in the Henon 
    map for arguments x, y with kwargs a and
    b as constants.
    '''
    x_next = x*(r-r*x)
    return x_next
    
# number of iterations and array initialization
steps = 10000
X = np.zeros(steps + 1)

# starting point
X[0] = 0.4

# add points to array
for i in range(steps):
    x_next = logistic_attractor(X[i])
    X[i+1] = x_next 

# plot figure
plt.plot(X[200:300], alpha = 0.8, markersize=0.3)
# plt.axis('off')
plt.show()
plt.close()
```

## Correlation dimension 

The logistic map is a strange attractor. Therefore, we can compute fractal dimension of the attractor basin via correlation dimension, which is expected to be 0.56 according to Procaccia et al. 1983

```python
sd_x= np.std(X, ddof=1)
rvals_vect = nolds.logarithmic_r(0.01 * sd_x, 2.0* sd_x, 1.03)
dim_corr,set_cr = nolds.corr_dim(X,emb_dim=1, rvals=rvals_vect, debug_data=True)
```

## Hurst exponent

Given the fact $D<1$, we can also estimate the Hurst exponent and assess its realtionship with the fractal dimension.

```python
nMin = 16
nMax = 2**12
nScales = 32
exponents = np.linspace(np.log2(nMin), np.log2(nMax), nScales)
scales = np.round(np.power(2.0, exponents)).astype(np.int64)
q_list = [-5, -3, -1, 1, 3, 5]
```

```python
def get_idxs(vec, scales):
    idxs = []
    for s in scales:
        idxs.append(np.where(vec==s)[0][0])
    return idxs
```

```python
def test_mat_dfa_lorenz(x):
    x_dfa = fathon.DFA(fu.toAggregated(x))
    n_w, F_w = x_dfa.computeFlucVec(scales, revSeg=False, polOrd=1)
    idxs = get_idxs(n_w, scales)
    n_w = n_w[idxs]
    F_w = F_w[idxs]
    H, _ = np.polyfit(np.log2(n_w), np.log2(F_w), 1)
    return H,n_w,F_w
```

```python
H,n_w,F_w = test_mat_dfa_lorenz(X)
print(H)
```

## Poincare Plot

The Poincare Plot is a simplified representation of the phase-space and it is obtained by plotting $X_{t+1}$ vs $X_{t}$. Similarly to the Hurst exponent or the sample entropy, one can compute the self-similarity of the attractor basin by deriving the variance along the two principal axis of the Poincare Plots. See the following reference for details related the heart-rate variability analysis:
https://journals.physiology.org/doi/full/10.1152/ajpheart.00405.2000

```python
X_lag_0 = X[:X.shape[0]-1]
X_lag_1 = X[1:]
plt.plot(X_lag_0,X_lag_1,'o')
```
